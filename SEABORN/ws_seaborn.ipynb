```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd  # for DataFrame

x=[1,2,3,4,5,6,7]
y=[20,40,30,40,30,50,30]

df=pd.DataFrame({"Days":x, "No of people":y})
df.head(7)
```

```python
#2 seaborn Lineplot
sns.lineplot(x="Days",y="No of people", data=df)
plt.show()
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
#this is for lineplot
sns.lineplot(x="day",y="total_bill",data=df,hue="sex",style="time",palette="flare")
plt.show()
```

```python
df1 = sns.load_dataset("titanic")
df1.head()
```

```python
# sns.barplot(x="who",y="fare",data=df1,color="magenta")
# plt.show()
sns.barplot(x="who",y="fare",data=df1,hue="sex")
plt.show()
```

```python
sns.barplot(x="who",y="fare",data=df1,palette="flare")
plt.show()
```

```python
sns.barplot(x="fare",y="who",data=df1,saturation=0.920,ci=False)
plt.show()
```

```python
from numpy import mean 
sns.barplot(x="who",y="fare",data=df1,estimator=mean)
plt.show()
```

```python
sns.barplot(x="who",y="fare",data=df1,facecolor=[1,1,1,0],edgecolor="0.2")
plt.show()
```
